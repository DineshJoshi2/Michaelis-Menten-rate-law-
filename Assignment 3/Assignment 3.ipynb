### Eadie_Hofstee Model

```python
import Eadie_Hofstee as EH
EH.Eadie_Hofstee_plot(df)
```

The equation for Eadie Hofstee plot is
$-r_{urea}$= $V_{max}$ - $K_m$*($-r_{urea}$/$C_{urea}$) which can be compared to linear equation y = mx +c to compute the value for $V_{max}$ and $K_m$

The Eadie–Hofstee plot does not
bias the points at low substrate
concentrations.

like other straight-line forms of the Michaelis–Menten equation, the Eadie–Hofstee plot was used historically for rapid evaluation of the parameters, but has been largely superseded by nonlinear regression methods that are significantly more accurate and no longer computationally inaccessible.

### Hane woolf model

```python
import Hanes_Woolf as HW
HW.Hanes_Woolf_plot(df)
```

The equation for the Hane woolf model is

$C_{urea}$/$-r_{urea}$ = $K_m$/$V_{max}$ + ($C_{urea}$/$V_{max}$). The value for $K_m$/$V_{max}$ is given by the Y intercept and the value for $1/V_{max}$ is given by the slope.

The Hanes–Woolf plot gives a
more accurate evaluation of
Vmax.

Like other techniques that linearize the Michaelis–Menten equation, the Hanes–Woolf plot was used historically for rapid determination of the kinetic parameters but it has been largely superseded by nonlinear regression methods that are significantly more accurate and no longer computationally inaccessible. It remains useful, however, as a means to present data graphically.

### Non Linear Regression

```python
import non_linear_regression as nlg
nlg.non_linear_regression_plot(df)
```

Nonlinear regression is a form of regression analysis in which observational data are modeled by a function which is a nonlinear combination of the model parameters and depends on one or more independent variables. The data are fitted by a method of successive approximations. Nonlinear regression is considered more accurate for calulating the michaelis menten kinetics parameters than other linearization methods.
